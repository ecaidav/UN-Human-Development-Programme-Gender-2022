```python
import pandas as pd
import altair as alt
from vega_datasets import data
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
inequality_index = pd.read_csv('/content/drive/MyDrive/Columbia University/Spring 2024/Careers/PassBlue/HDR23-24_Statistical_Annex_GII_Table.csv')
```

```python
inequality_index['HDI rank'] = pd.to_numeric(inequality_index['HDI rank'], errors='coerce')
inequality_index = inequality_index.dropna(subset=['HDI rank'])
inequality_index['HDI rank'] = inequality_index['HDI rank'].astype(int)

inequality_index['Share of seats in parliament (% held by women)'] = pd.to_numeric(inequality_index['Share of seats in parliament (% held by women)'], errors='coerce')
inequality_index = inequality_index.dropna(subset=['Share of seats in parliament (% held by women)'])
inequality_index['Share of seats in parliament (% held by women)'] = inequality_index['Share of seats in parliament (% held by women)'].astype(int)
```

```python
inequality_index.head()
```

```python
countries = inequality_index[inequality_index['Type'] == 'Country']
```

```python
inequality_index['Gender Inequality Index: Rank'] = pd.to_numeric(inequality_index['Gender Inequality Index: Rank'], errors='coerce')
gender_index = inequality_index[inequality_index['Gender Inequality Index: Rank'].notna()]
gender_index['Gender Inequality Index: Rank'] = gender_index['Gender Inequality Index: Rank'].astype('Int64')
gender_index['HDI_Gender'] = gender_index['HDI rank'] - gender_index['Gender Inequality Index: Rank']
```

```python
gender_index = gender_index.rename(columns={
    'Maternal mortality ratio (deaths per 100,000 live births)': 'Maternal Mortality Ratio',
    'Asolescent birth rate (births per 1,000 women ages 15â€“19)': 'Adolescent Birth Rate',
    'Share of seats in parliament (% held by women)': 'Parliament Seats Held by Women',
    'Population with at least some secondary education - % Females ages 25 and older': 'Women with Some Secondary Education',
    'Population with at least some secondary education - % Males ages 25 and older': 'Men with Some Secondary Education',
    'Labour force participation rate - % Females ages 15 and older': 'Women in the Labor Force',
    'Labour force participation rate - % Males ages 15 and older': 'Men in the Labor Force',
    'Gender Inequality Index: Value': 'Gender Inequality Index',
    'Gender Inequality Index: Rank': 'GII Rank'
})
```

```python
columns_to_percentage = [
    'Parliament Seats Held by Women',
    'Women with Some Secondary Education',
    'Men with Some Secondary Education',
    'Women in the Labor Force',
    'Men in the Labor Force'
]

for column in columns_to_percentage:
    gender_index[column] = pd.to_numeric(gender_index[column], errors='coerce') / 100.0

gender_index[columns_to_percentage] = gender_index[columns_to_percentage].apply(lambda x: x * 100)

for column in columns_to_percentage:
    new_column_name = column + ' %'
    gender_index[new_column_name] = pd.to_numeric(gender_index[column], errors='coerce')
```

```python
other_columns_to_format = [
    'Parliament Seats Held by Women %',
    'Women with Some Secondary Education %',
    'Men with Some Secondary Education %',
    'Women in the Labor Force %',
    'Men in the Labor Force %'
]

for column in other_columns_to_format:
    gender_index[column] = pd.to_numeric(gender_index[column.replace(' %', '')], errors='coerce')
    gender_index[column] = gender_index[column].apply(
        lambda x: f"{int(x)}%" if pd.notnull(x) and x.is_integer() else f"{x:.1f}".rstrip('0').rstrip('.') + "%" if pd.notnull(x) else "NaN"
    )
```
